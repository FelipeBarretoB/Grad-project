## Modelado
Dado que queremos comparar modelos con los diferentes datos, vamos a usar pycaret para crear varios modelos de clasificación y compararlos.

### Internet Firewall

```python
#Configuramos el entorno de pycaret, usamos session id para hacer datos reproducibles y use_gpu para acelerar el proceso, dado que se cuenta con una GPU
inf_clf = setup(data = inf, target = 'Action', session_id = 123, use_gpu=True)
```

Creamos los modelos

```python
#Creamos los modelos
inf_clf_models = compare_models()
```

Guardamos el mejor modelo

```python
#Guardamos el mejor modelo
inf_best_model = inf_clf_models[0]
evaluate_model(inf_best_model)
```

```python
print(inf_best_model)
```

```python
predict_model(inf_best_model, data=inf)
```

Con esto, concluimos el modelado de los datos de Internet Firewall, ahora vamos a realizar el modelado de los otros datasets

### KDD Cup 99

En este caso, podemos entrenar el modelo de dos maneras, binaria o multiclase, en este caso vamos a entrenar el modelo de manera binaria

```python
kdd_binary = kdd.drop(columns=['attack', 'Attack Type'])
kdd_binary.corrwith(kdd['Is attack']).abs().sort_values(ascending=False)
```

```python
setup(data = kdd_binary, target = 'Is attack', session_id = 123, use_gpu=True)
```

Creamos los modelos

```python
kdd_clf_models = compare_models()
```

Guardamos el mejor modelo
Aunque es muy bueno, deberíamos tener en cuenta que puede estar sobre entrenado

```python
#Guardamos el mejor modelo
kdd_best_model = kdd_clf_models
evaluate_model(kdd_best_model)
```

```python
predict_model(kdd_best_model, data=kdd_binary)
```

Ahora vamos a entrenar el modelo de manera multiclase

Borramos las columnas que no necesitamos

```python
kdd_multiclass = kdd.drop(columns=['Is attack', 'attack'])
kdd_multiclass.corrwith(kdd['Attack Type']).abs().sort_values(ascending=False)
```

```python
setup(data = kdd_multiclass, target = 'Attack Type', session_id = 123, use_gpu=True)
```

Creamos los modelos

Guardamos el mejor modelo
