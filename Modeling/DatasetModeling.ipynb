```python
from pycaret.classification import *
```

```python
duplicate_indices = final_df[final_df.index.duplicated()]
print(duplicate_indices.sum())
```

```python
final_df = final_df[~final_df.index.duplicated()]
final_df.reset_index(drop=True, inplace=True)
```

```python
import numpy as np
# Reemplazar infinitos con NaN
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Eliminar filas con NaN
final_df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
##scaler = MinMaxScaler()
## df_normalized = pd.DataFrame(scaler.fit_transform(final_df), columns=final_df.columns)

target = final_df['Target']
features = final_df.drop(columns=['Target'])

# Aplicar la normalizaciÃ³n solo a las caracterÃ­sticas
scaler = MinMaxScaler()
features_normalized = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)

# Concatenar las caracterÃ­sticas normalizadas con la variable objetivo
final_df_normalized = pd.concat([features_normalized, target], axis=1)
```

```python
final_df_normalized['Target'].value_counts()
```

```python
final_df_normalized.dropna(subset=['Target'], inplace=True)
```

```python
setup(final_df_normalized, target='Target')
```

```python
compare_models(errors = 'raise')
```

```python
best_model = compare_models(errors = 'raise')
```

```python
# Ajuste del modelo seleccionado
final_model = create_model(best_model)

# EvaluaciÃ³n del modelo
evaluate_model(final_model)
```
